```python
#-*- coding:utf-8 -*-
import requests, os

# pip install beautifulsoup4
import bs4
from bs4 import BeautifulSoup
from urllib import parse
import re
import lxml
import json
import sys

# import selenium.webdriver as webdriver
# import chardet
# https://www.codeit.kr/questions/1205
```

```python
# 읽어온 html파일 저장
def save_html(soup):
    soup_string = str(soup.html)
    #print(soup)

    file_name = word + "_html.txt"
    with open(file_name, 'w', encoding='UTF-8') as f:
        f.write(soup_string)
        f.close()
    
# 페이지 주소 리스트 획득
def get_shortcode_list(soup):
    soup_script = soup.find_all("script")[3].string
    soup_script = re.split(':|,|\"|\'|\{|\}|\[|\]',soup_script)
    shortcodes = []
    temp = 0
    while 1:
        if 'shortcode' in soup_script[int(temp):]:

            new_shortcode = soup_script.index('shortcode', temp) + 3
            shortcodes.append(soup_script[new_shortcode])
            temp = new_shortcode + 1
        else :
            print("Get ", len(shortcodes), " shortcodes")
            return shortcodes

def is_the_only_string_within_a_tag(s):
    """Return True if this string is the only child of its parent tag."""
    return (s == s.parent.string)
        
# 텍스트 획득
def get_text(page):
    text_script = page.find_all("script")[3].string # return <class 'bs4.element.Tag'>
    # text_script = page.find_all(string=is_the_only_string_within_a_tag)
    # print(text_script)
    
    # text_script = text_script.string
    # text_script = BeautifulSoup.unicode(page.find_all("script")[3])
    text_script = re.split(':|,|\"|\'|\{|\}|\[|\]',text_script)
    text_index = text_script.index('text') + 3
    text = text_script[text_index]
    
    return text   
```

```python
# word = input("\n** 원하는 태그를 입력하세요: \n>>> ")
word = "경복궁"
```

```python
# 한글 포함 URL열기
url_tmp = "www.instagram.com/explore/tags/" + word
url = "http://" + parse.quote(url_tmp)
#url = "http://www.instagram.com/explore/tags/%EB%82%A8%EC%82%B0"
print(url)

# html 파일 읽음
response = requests.get(url).text
soup = BeautifulSoup(response, 'lxml')

# 파싱하여 shortcode 가 포함된 <script>획득
shortcodes = get_shortcode_list(soup)
```

```python
i = 1
json_data = []

for shortcode in shortcodes:
    print(i, "th ", shortcode)
    i = i + 1  
    
    try:
        page_url = "https://www.instagram.com/p/" + shortcode + "/"
        # do not use 'response = requests.get(page_url).text'
        response = requests.get(page_url).content
        page = BeautifulSoup(response, 'lxml')
        
        for tag in page.select('meta[property*=image]'):
            image_url = tag['content']

        for tag in page.select('meta[property*=og:description]'):
            description = tag['content']

        # 파싱하여 text 가 포함된 <script>획득
        text = get_text(page)
        text = text.replace("\\n", "\\r\\n")
        # \u000d
        # print("text: (", unicode(text), ")")

        # 요약된 텍스트 분리
        # print("description: ", description)
        # print("desciprtion ", type(description))
        data = re.split('on Instagram:', description)

        # print("text: ", text)

        # print("data[0]: ", data[0])
        if '(' in data[0]:
            user_id_index = 6
        else:
            user_id_index = 5

        # print("user_id_index : ", user_id_index)

        data = re.split(",|\(|\)|@|Likes|Comments|-",data[0])
        # print("data: ", data)

        likes = data[0]
        # print("likes: ", likes)
        user_id = data[user_id_index]
        # print("user_id: ", user_id)
        img_name = word + "_" + shortcode + ".jpg"

        output = {"shortcode":shortcode, "label":word, "likes":likes, "user_id":user_id, "text":text}
        json_data.append(output)  
        
        # 사진 raw 파일과 파일명 받기
        pic_response = requests.get(image_url).content
        file_name = "./img/" + img_name
        # 사진 저장하기
        with open(file_name, 'wb') as f:
            f.write(pic_response)
        print("Okay")
        print()
    except:
        print("Unexpected error:", sys.exc_info()[0])
        print("Not okay")
        print()
        
        
if not os.path.isfile('label.json'):
    with open('label.json', mode='w', encoding='UTF-8-sig') as f:
        f.write(json.dumps(json_data, ensure_ascii=False, indent=2))
else:
    with open('label.json') as feedsjson:
        feeds = json.load(feedsjson)

    feeds.append(json_data)
    with open('label.json', mode='w', encoding='UTF-8-sig') as f:
        f.write(json.dumps(feeds, ensure_ascii=False, indent=2))
        
print("json file saved")
```
